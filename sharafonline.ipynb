```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import json
```

```python
urls = ['https://sharafonline.ir/product-sitemap.xml',
        'https://sharafonline.ir/product-sitemap2.xml',
        'https://sharafonline.ir/product-sitemap3.xml',
        'https://sharafonline.ir/product-sitemap4.xml',
        'https://sharafonline.ir/product-sitemap5.xml',
        'https://sharafonline.ir/product-sitemap6.xml',
        'https://sharafonline.ir/product-sitemap7.xml']

links = pd.DataFrame()
```

```python
for s in urls:
    xml = pd.read_xml(s)
    print(len(xml))
    links = links.append(xml)
    print(len(links))
```

```python
links
```

```python
product = []
```

```python
for n in range(1,252):
    url = f'https://sharafonline.ir/shop/page/{n}/'
    page = r.get(url)

    print(f"{str(url)} is browesd")
    soup = bs(page.content, 'html5lib')
    as_ = soup.find_all("a", class_ = "product-image-link")

    for href in as_:
        product.append(href["href"])
    print(f'{str(url)} is done crawling. Total product till now {len(product)}')
```

```python
product
```

```python
contents = pd.DataFrame()
```

```python
for i, product_url in enumerate(product[5461:]):

    test = r.get(product_url)

    test_soup = bs(test.content, 'html5lib')

    graph = test_soup.find("script" , class_="yoast-schema-graph")

    print(f"{i+1} opend")

    jgragh = json.loads(graph.string)["@graph"][0]
    g_url = jgragh.get("url")
    g_name = jgragh.get("name")
    g_published = jgragh.get("datePublished")
    g_modified = jgragh.get("dateModified")
    g_description = jgragh.get("description")

    price = test_soup.find("p", class_="price").text.split()
    if len(price) > 2:
        g_main = price[0]
        g_disc = price[2]
        sold = 0
    elif 0 < len(price):
        g_main = price[0]
        g_disc = ""
        sold = 0
    else:
        g_main = "این محصول فروخته شده است"
        g_disc = "این محصول فروخته شده است"
        sold = 1

    contents = contents.append({"لینک":g_url,
                                "عنوان":g_name,
                                "تاریخ انتشار":pd.to_datetime(g_published).tz_localize(None),
                                "تاریخ آخرین اصلاح":pd.to_datetime(g_modified).tz_localize(None),
                                "توضیحات":g_description,
                                "قیمت اصلی":g_main,
                                "قیمت با تخفیف":g_disc,
                                "فروخته شده":sold}, ignore_index=True)
    print(f"{i+1}/{len(product[5461:])} is done")
```
