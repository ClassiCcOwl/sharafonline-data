```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import json
```

```python
urls = ['https://sharafonline.ir/product-sitemap.xml',
        'https://sharafonline.ir/product-sitemap2.xml',
        'https://sharafonline.ir/product-sitemap3.xml',
        'https://sharafonline.ir/product-sitemap4.xml',
        'https://sharafonline.ir/product-sitemap5.xml',
        'https://sharafonline.ir/product-sitemap6.xml',
        'https://sharafonline.ir/product-sitemap7.xml']

links = pd.DataFrame()
```

```python
for s in urls:
    xml = pd.read_xml(s)
    print(len(xml))
    links = links.append(xml)
    print(len(links))
```
